```python
import pandas as pd

data = {
    "Year": [1970,1971,1972,1973,1974,1975,1976,1977,1978,1979,
             1980,1981,1982,1983,1984,1985,1986,1987,1988,1989,1990],
    "Income": [3399,3768,4251,4804,5576,5998,7062,7100,7240,7600,
               8000,8200,8300,8600,9000,9400,9800,10200,10600,11000,11500]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Year"]]
y = df["Income"]

model = LinearRegression()
model.fit(X, y)

print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
future_year = [[2025]]
prediction = model.predict(future_year)

print("Predicted income for 2025:", prediction[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color="blue")
plt.plot(X, model.predict(X), color="red")
plt.xlabel("Year")
plt.ylabel("Canada Income")
plt.title("Linear Regression - Canada Income Prediction")
plt.show()
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)
print("R² Score:", r2_score(y, y_pred))
```
