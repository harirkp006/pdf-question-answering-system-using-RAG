```python
import pandas as pd

data = {
    "Year": [1970, 1971, 1972, 1973, 1974, 1975],
    "GDP": [2000, 2100, 2200, 2400, 2600, 2800],
    "Population": [30, 31, 32, 33, 34, 35],
    "Income": [3399, 3768, 4251, 4804, 5576, 6000]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression
X = df[["Year", "GDP", "Population"]]
y = df["Income"]
model = LinearRegression()
model.fit(X, y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
future_data = pd.DataFrame([[2025, 5000, 40]],
                           columns=["Year", "GDP", "Population"])

prediction = model.predict(future_data)

print("Predicted Income for 2025:", prediction[0])
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)

print("R² Score:", r2_score(y, y_pred))
```
